# plot을 이용한 그래프 그리기

## 맷플롯립을 이용한 그래프 생성

```python
x_numbers = [1, 2, 3] # x 좌표 리스트
y_numbers = [2, 4, 6] # y 좌표 리스트
```

```python
from pylab import plot, show
plot(x_numbers, y_numbers)
```

## 그래프에 점 표시

```python
plot(x_numbers, y_numbers, marker='o') # marker인자를 사용하여 점 표시
```

```python
plot(x_numbers, y_numbers, marker='*') # o * x  + 등이 있음
```

```python
plot(x_numbers, y_numbers, 'o') # 각각의 값을 점으로만 표시
```

## 수식으로 그래프 그리기

### 만유인력에 대한 뉴턴의 법칙

```python
'''
두 물체간 거리와 인장력 간 관계
'''

import matplotlib.pyplot as plt

#그래프 그리기

def draw_graph(x,y):
    plt.plot(x, y, marker='o')
    plt.xlabel('Distance in meters')
    plt.ylabel('Gravitational force in newtons')
    plt.title('Gravitational force and distance')
    plt.show()
    
def generate_F_r():
    # r에 대한 값을 생성하기
    r = range(100, 1001, 50)
    # F의 값을 저장하기 위한 공백 리스트
    F = []
    
    # 상수, G
    G = 6.674*(10**-11)
    # 두 개의 질량값
    m1 = 0.5
    m2 = 1.5
    
    # 힘을 계산하고 리스트 F에 추가한다.
    
    for dist in r:
        force = G*(m1*m2)/(dist**2)
        F.append(force)
        
    # draw_graph 함수를 호출한다.
    draw_graph(r, F)
    
if __name__ =='__name__':
    generate_F_r()
```
